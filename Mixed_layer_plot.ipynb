## Vertical velocity with only values inside the cooling disk

```python
%%time
vertical_velocity1 = ds1['w'][:,:128,:,:].where(radius<=1,0)
vertical_velocity1
```

```python
%%time
vertical_velocity1.isel(time=20,zF = 127).plot()
```

```python
ds1['zF'][127]
```

```python
%%time
vertical_velocity2 = ds2['w'][:,:128,:,:].where(radius<=1,0)
vertical_velocity3 = ds3['w'][:,:128,:,:].where(radius<=1,0)
vertical_velocity4 = ds4['w'][:,:64,:,:].where(radius<=1,0)
vertical_velocity5 = ds5['w'][:,:64,:,:].where(radius<=1,0)
```

## Modified temperature

```python
modified_temp1 = (ds1['T']-ds1['T'].isel(time=0)).where(radius<=1,0)
modified_temp1.isel(time=65,xC = 128).plot()
```

```python
%%time
modified_temp2 = (ds2['T']-ds2['T'].isel(time=0)).where(radius<=1,0)
modified_temp3 = (ds3['T']-ds3['T'].isel(time=0)).where(radius<=1,0)
modified_temp4 = (ds4['T']-ds4['T'].isel(time=0)).where(radius<=1,0)
modified_temp5 = (ds5['T']-ds5['T'].isel(time=0)).where(radius<=1,0)
```

## Calculating Buoyancy flux at each position

```python
%%time
buo1 = vertical_velocity1*modified_temp1*g*alpha
buo1
```

```python
%%time
buo2 = vertical_velocity2*modified_temp2*g*alpha
buo3 = vertical_velocity3*modified_temp3*g*alpha
buo4 = vertical_velocity4*modified_temp4*g*alpha
buo5 = vertical_velocity5*modified_temp5*g*alpha
```

## Average Vertical buoyancy flux 

```python
avg_buoyancy1 = (buo1.sum(dim = 'xC').sum(dim = 'yC'))/12851
avg_buoyancy1
```

```python
%%time
avg_buoyancy2 = (buo2.mean(dim = 'xC').mean(dim = 'yC'))/12851
avg_buoyancy3 = (buo3.mean(dim = 'xC').mean(dim = 'yC'))/12851
avg_buoyancy4 = (buo4.mean(dim = 'xC').mean(dim = 'yC'))/12851
avg_buoyancy5 = (buo5.mean(dim = 'xC').mean(dim = 'yC'))/12851
```

## accessing their diagonal element

```python
(np.diagonal(avg_buoyancy1.values, offset=0, axis1=1, axis2=2)).shape
```

```python
%%time
avg_buoyancy1 = np.diagonal(avg_buoyancy1.values, offset=0, axis1=1, axis2=2)
avg_buoyancy2 = np.diagonal(avg_buoyancy2.values, offset=0, axis1=1, axis2=2)
avg_buoyancy3 = np.diagonal(avg_buoyancy3.values, offset=0, axis1=1, axis2=2)
avg_buoyancy4 = np.diagonal(avg_buoyancy4.values, offset=0, axis1=1, axis2=2)
avg_buoyancy5 = np.diagonal(avg_buoyancy5.values, offset=0, axis1=1, axis2=2)
```
